```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sales.csv')
df
```

```python
df.drop(['Status', 'unnamed1'], axis=1,inplace=True)
df
```

## People of which field buy our products the most

```python
occupation1 = df.Occupation
occupation = pd.DataFrame(occupation1)
occupation = df.pivot_table(index = ['Occupation'], aggfunc='size')
occupation
```

```python
occupation.plot(kind='pie', autopct='%0.1f%%',radius=2)
plt.show()
```

## Which state should we focus more on?

```python
state1 = df.State
state = pd.DataFrame(state1)
state = df.pivot_table(index = ['State'], aggfunc='size')
state
```

```python
state.plot(kind='pie', autopct='%0.1f%%', radius=2.5)
plt.show()
```

## Which Age Group buys our products the most?

```python
age_group_1 = df.Age_Group
age_group = pd.DataFrame(age_group_1)
age_group = df.pivot_table(index = ['Age_Group'], aggfunc='size')
age_group
ag_values = (age_group.values)
ag_index = np.array(age_group.index)
```

```python
plt.bar(ag_index, ag_values/ 10**2,color='red')
plt.xlabel('Age Group')
plt.ylabel('Percentage(%)')
plt.show()
```

## Which zone is highely enthusiastic?

```python
zone1 = df.Zone
zone = pd.DataFrame(zone1)
zone = df.pivot_table(index=['Zone'], aggfunc='size')
z_index = np.array(zone.index)
z_values = np.array(zone.values)
```

```python
plt.bar(z_index, z_values)
plt.xlabel('Zones')
plt.ylabel('Amount of Products Sold')
for index, value in enumerate(z_values):
    plt.text(index, value, str(value))
```

## Which Types of Products are usually bought?

```python
product_category1 = df.Product_Category
product_category = pd.DataFrame(product_category1)
product_category = df.pivot_table(index=['Product_Category'], aggfunc='size')
product_category1

pc_index = np.array(product_category.index)
pc_values = np.array(product_category.values)
```

```python
plt.figure(figsize=(15,6))
plt.plot(pc_index, pc_values, marker='.', color='orange', linewidth='2')
plt.xticks(rotation='vertical', size='8')
for index, value in enumerate(pc_values):
    plt.text(index, value, str(value))
plt.show()
```
